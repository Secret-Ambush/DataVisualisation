**Feature Selection**
1. Univariate Selection
2. Feature Importance
3. Correlation Matrix

```python
x = data.drop(["price_range"],axis=1)
y = data["price_range"]
```

```python
## Feature Importance
```

```python
from sklearn.ensemble import ExtraTreesClassifier
model = ExtraTreesClassifier()
model.fit(x,y)
df = pd.DataFrame()
df['Features'] = x.columns
df['Scores'] = model.feature_importances_
df = df.sort_values('Scores',ascending=False).reset_index()
df = df.drop('index',axis=1)
df
```

```python
# Univariate Selection
```

```python
from sklearn.feature_selection import SelectKBest,chi2
best_features = SelectKBest(k=10,score_func=chi2)
model = best_features.fit(x,y)
df = pd.DataFrame()
df['Features'] = x.columns
df['Scores'] = model.scores_
df = df.sort_values('Scores',ascending=False).reset_index()
df = df.drop('index',axis=1)
df
```

```python
#Correlation Matrix
```

```python
import numpy as np
corr = data.corr()
corr = corr['price_range']
corr = np.abs(corr)
corr = corr.sort_values(ascending=False)[1:] #this means skip first row
corr
```

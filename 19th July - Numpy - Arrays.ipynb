**Array**
1. It is a group of homogenous data (similar data)
2. It can be multidimensional
3. It is preferred over list or tuple in case of doing mathematical operation
4. Numpy (Numerical Python) - It is a library/package that can be used to work on mutlidimensional arrays and apply different mathematical operations


```python
import numpy as np
```

```python
a1 = [10,20,30]
a2= [ 1,2,3]
print(a1 + a2) #addition on lists = add strings
print(a1 * 3)  #repeat thrice
```

```python
sum = []
for i in range (len(a1)):
  sum.append(a1[i]+a2[i])
print(sum)
```

```python
a3 = np.array([10,20,30])
a4 = np.array([1,2,3])
print(a3 + a4)  #addition
print(a3 * 3)   #multiplication
```

```python
np.random.seed(1)
data = np.random.randint(100,500,(5,6))
print(data)
```

```python
print(data.min())
print(data.min(axis=0))   ## column wise
print(data.min(axis=1))   ## row wise
```

```python
print(data.max())
print(data.max(axis=0))   ## column wise
print(data.max(axis=1))   ## row wise
```

```python
print(data.sum())
print(data.sum(axis=0))   ## column wise
print(data.sum(axis=1))   ## row wise
```

```python
print(data.mean())
print(data.mean(axis=0))   ## column wise
print(data.mean(axis=1))   ## row wise
```

```python
print(data.argmax())
print(data.argmax(axis=0))   ## column wise
print(data.argmax(axis=1))   ## row wise
```

```python
print(data.argmin())
print(data.argmin(axis=0))   ## column wise
print(data.argmin(axis=1))   ## row wise
```

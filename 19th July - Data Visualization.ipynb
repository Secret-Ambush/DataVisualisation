```python
### Histogram (Frequency Distribution)
np.random.seed(1)
data = np.random.randint(1,100,50)
print(data)
plt.hist(data,rwidth=0.9,bins=5)
plt.show()
```

```python
### Box Plot (De JVC tect Outliers)
samples = [10,20,12,25,50,34,23,40,45,60,22,21,34,23,33,
           10,20,12,25,50,34,23,40,45,60,22,21,34,23,33,
           10,20,12,25,50,34,23,90,45,60,22,21,34,23,33,
           10,20,100,25,50,34,23,40,45,60,22,21,34,23,33,
           10,20,12,25,50,34,23,40,45,60,22,21,34,23,33,
           10,20,12,25,50,34,23,40,45,60,22,21,34,23,33,-100,-90]
sns.boxplot(x=samples)
plt.show()
```

```python
salary = [20000,40000,30000,30000,50000]
years = [1,3,2,2,4]
gender = ["M","F","F","M","M"]
role = ["Developer","Manager","Designer","Developer","Manager"]
```

```python
sns.countplot(x=gender)
plt.show()
```

```python
sns.countplot(x=role)
plt.show()
```

```python
sns.countplot(x=gender,hue=role)
plt.show()
```

```python
sns.countplot(x=role,hue=gender)
plt.show()
```

```python
### Bivariate
```

```python
### Line Plot
plt.plot(years,salary)
plt.show()
```

```python
### Scatter Plot
plt.scatter(years,salary, s= 100)
plt.show()
```

```python
### Bar Plot
plt.bar(years,salary, color ='green')
plt.xticks(years)
plt.show()
```

```python
### Relplot
sns.relplot(x= years, y= salary, hue = gender, s = 100)
plt.show()
```

```python
## relplot
sns.relplot(x=years,y=salary,hue=role,s=300)
plt.show()
```

```python
sns.get_dataset_names()
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
sns.relplot(x="total_bill",y="tip",data=tips,s=100,hue='sex',col="sex",col_wrap=2)
plt.show()
```

```python
sns.relplot(x="total_bill",y="tip",data=tips,s=100,hue='day',col="day",col_wrap=2)
plt.show()
```

```python
### Jointplot (Scatter Plot + Histogram)
sns.jointplot(x='total_bill',y='tip',data=tips)
plt.show()
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg')
plt.show()
```

```python
### Multivariate
a = sns.PairGrid(tips)
a.map_diag(sns.histplot)
a.map_offdiag(sns.scatterplot)
plt.show()
```

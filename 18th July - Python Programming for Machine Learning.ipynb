**List**
1. Group of heterogenous data (mix of data)
2. It is mutable (which can be modified)
3. It is identified by [a1,a2,a3,a4] 

**Tuple**
1. Group of heterogenous data (mix of data)
2. It is immutable (which cannot be modified)
3. It is identified by (a1,a2,a3,a4)

**Dictionary**
1. It stores the data in the form of key-value pairs, where key is always unique
2. Key can be a string and value can be anything
3. Identified by {key1 : value1, key2: value2 ...}

**Sets**
1. Unordered collection of data 
2. Data is always unique
3. Identified by {a1,a2,a3...}


```python
a1 = []
#a1 = list
print(type(a1))
a1 = [10,34.3,"Umang",True]
print(a1)
```

**Conditional Statement**

```python
num = int(input("Enter Number:"))
if num > 0:
  print("Positive number")
elif num == 0:
  print("Zero")
else:
  print("Negative number") #use tab for indents
```

```python
try: 
  num1 = int(input("Enter First Number:"))
  num2 = int(input("Enter Second Number:"))
  op = input("Enter Operation(+,-,*,/): ")
  if op == "+":
    print("Result is",num1 + num2)
  elif op == "-":
    print("Result is",num1 - num2)
  elif op == "*":
    print("Result is",num1 * num2)
  elif op == "/":
    print("Result is",num1 / num2)
  else:
    print("Enter valid operator")

except ZeroDivisionError:
  print("----Cannot divide by zero")
```

**Functions** - to reuse code
1. Function Without Argument
2. Function With Argument
3. Function With Default Argument
4. Function With return statement

```python
## Function Without Argument
def add_nums():
  print(10+20)

add_nums()
```

```python
## Function with non default Argument
def sub_nums(a,b):
  print(a-b)

sub_nums(10, 20)
```

```python
def sub_nums(num1,num2):
  print(num1-num2)

sub_nums(10, 20)
```

```python
## Function with default argument
def calc(num2,num1= 10,op="+"):
  if op == "+":
    result = num1 + num2
  elif op == "-":
    result = num1 - num2
  elif op == "*":
    result = num1 * num2
  else:
    result = num1/num2
  print("The result is ",result)
```

```python
## function with return variable
def calc_r(num1,num2,op="+"):
  '''
  This function is doing the calculation of two numbers where additon is by default
  (documentation)
  '''
  if op == "+":
    result = num1 + num2
  elif op == "-":
    result = num1 - num2
  elif op == "*":
    result = num1 * num2
  else:
    result = num1/num2
  return result
```
